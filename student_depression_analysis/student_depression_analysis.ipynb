```python
## Drop columns which does not have much correlation with target column
df.drop(columns=['City','Degree','Profession'],inplace=True)
```

## Split the data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Depression']),df['Depression'],
                                            test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

## Data Transformation

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# one hot encoding
trf1 = ColumnTransformer([
    ('ohe_gender_isSuicidal_isHistory', OneHotEncoder(handle_unknown='ignore'),[0,7,10])
],remainder='passthrough')
```

```python
from sklearn.preprocessing import OrdinalEncoder

trf2 = ColumnTransformer([
    ('oe_sleep_diet',OrdinalEncoder(categories=[['Others','Less than 5 hours','5-6 hours','7-8 hours','More than 8 hours'],
                                               ['Others','Unhealthy','Moderate','Healthy']],handle_unknown='use_encoded_value',unknown_value=-1),[10,11])
],remainder='passthrough')
```

```python
from sklearn.preprocessing import MinMaxScaler
# Scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,20))
])
```

## Model

```python
from sklearn.tree import DecisionTreeClassifier
# train the model
trf4 = DecisionTreeClassifier()
```

## Create Pipeline

```python
from sklearn.pipeline import Pipeline,make_pipeline

pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps
```

```python
y_pred = pipe.predict(X_test)
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### Cross validation

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

### GridSearch

```python
# gridsearchcv
params = {
    'trf4__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```
